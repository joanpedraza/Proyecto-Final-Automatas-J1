```python
perso.validate_input(respuestas_usuario, step=False)
```

```python
def run_dtm(dtm, input_str):
    dtm.initialize(input_str)
    while not dtm.input_tape.at_final_state():
        dtm.do_transition()
    return ''.join(dtm.input_tape.tape)
```

```python
output_str = run_dtm(perso, respuestas_usuario)
print(output_str)
```

```python
output_str = ''
for symbol in perso.validate_input(respuestas_usuario):
    output_str += perso.tape.read()
    perso.do_transition(symbol)

output_str += perso.tape.read()

print(respuestas_usuario)
print(f'Salida: {output_str}')
```
